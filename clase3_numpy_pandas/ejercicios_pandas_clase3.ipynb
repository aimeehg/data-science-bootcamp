```
"""
                                        EJERCICIOS PARA PRACTICAR
                                        Nombre: Aimee Hernandez
"""
import numpy as np
import pandas as pd

df = pd.read_csv('usersv2.csv')

# 1 Obtener el nombre y correo de todos los usuarios.

df[['name', 'email']]
```

```
# 2 Obtener el nombre de todos los usuarios cuyo sexo es Femenino.

df [df['gender'] == 'female']['name']
```

```
# 3 Obtener el nombre de todos los usuarios cuyo sexo es Masculino y tengan una
# edad mayor a 50

# Condición para btener los usuarios de sexo masculino y que su edad sea mayor a 50
condition = (df['gender'] == 'male') & (df['age'] > 50)
# Obtener los nombres de los usuarios respecto a la condición dada
df [condition]['name']
```

```
# 4 Obtener todos los usuarios cuya edad sea mayor a 68.

df [df['age'] > 68]
```

```
# Obtener el nombre y el pais del usuario mas joven

df.sort_values('age').head(1)[['name', 'country']]
```

```
# 5 Obtener el username y correo electrónico de los usuarios cuya edad se encuentre
# en el rango 50 a 60.

df [(df['age'] > 50) & (df['age'] < 60)][['username', 'email']]
```

```
# 6 Obtener el username de todos los usuarios cuyo correo no terminen con
# @example.com 
# como lo hice al principio
df[df['email'].str.endswith('@example.com') == -1 ]['username']
# como lo hizo Eduardo en clase
df[~ df['email'].str.endswith('@example.com') ]['username']

```

```
# 7 Obtener el username de todos los usuarios cuyo país sea Germany, Finland o Canada.
df[ (df['country'] == 'Germany') | (df['country'] == 'Finland') | (df['country'] == 'Canada')]['username']
# otra mejor manera
df [df['country'].isin(['Germany', 'Finland', 'Canada'])]['username']
```

```
# 8 Obtener el nombre y correo electronico de todos los usuarios de sexo Femenino
# que sean del país Alemania
df[ (df['gender'] == 'female') & (df['country'] == 'Germany') ][['name', 'email']]
```

```
# 9 Obtener el promedio de edad de todos los usuarios de sexo Femenido de Canada
# mayores de 20 años.

# Obtiene los usuarios de sexo femenino de Canadá mayores de 20 años 
condition = (df['gender'] == 'female') & (df['country'] == 'Canada') & (df['age'] > 20 )
# Obtiene el promedio de edad de los usuarios que cumplen la condición dada
df[condition]['age'].mean().round(2)
```

```
# 10 Conocer la cantidad de usuarios que residen en Finlandia.
df[ (df['country'] == 'Finland') ].shape[0]
```

```
# 11 Mostrar en consola la cantidad de hombre y mujeres en el dataframe.

#esta manera muestra los resultados separados
print(df[ (df['gender'] == 'male') ].shape[0])
print(df[ (df['gender'] == 'female') ].shape[0])

# una mejor manera con groupby

df.groupby('gender')['gender'].count()
```

```
# 12 Mostrar en consola el país con más mujeres

df [df['gender']=='female'].groupby('country')['country'].count().sort_values(ascending=False).head(1)
```

```
# 13 Obtener los 3 países con más usuarios.

df.groupby('country')['country'].count().sort_values(ascending=False).head(3)
```

```
# 14 Mostrar en consola el nombre de todos los usuarios cuya edad se encuentre en el
# rango de 10 a 20 y 40 a 70.
df[(df['age'].between(10, 20)) | (df['age'].between(40, 70))][['name','age']]
```

```
# 15 Mostrar en consola todos los usuarios con correo electrónico.
df[df['email'].notna()]
```

```
# 16 Mostrar en pantalla el nombre y correo electrónico del usuario más joven en Canadá.
df[df['country'] == 'Canada'].sort_values('age').head(1)[['name','email', 'age', 'country']]
```

```
# 17 Mostrar en pantalla el nombre y correo electrónico del usuario más viejo en Canada.
df[df['country'] == 'Canada'].sort_values('age', ascending=False).head(1)[['name','email', 'age', 'country']]
```

```
# 18 Listar en consola los 3 países con menor cantidad de usuarios
df.groupby('country')['country'].count().sort_values().head(3)
```

```
# 19 Obtener el país con mayor cantidad de usuarios cuya edad sea mayor a 50.
df[df['age'] > 50].groupby('country')['country'].count().sort_values(ascending=False).head(1)
```

```
# 20 Obtener el país con mayor promedio de edad.
df.groupby('country')['age'].mean().round(2).sort_values(ascending=False).head(1)
```

```
# 21 Mostrar en consola el país con más hombres
df[df['gender'] == 'male'].groupby('country')['country'].count().sort_values(ascending=False).head(1)
```

```
# 22 Mostrar en consola el nombre, username y edad de todos los usuarios cuya edad
# es mayor a 10 y no sean del país México, Brasil y Canadá.

# Obtiene los usuarios de edad mayor a 10 que no sean de México, Brasil y Canadá
condition = (df['age'] > 10) & (~df['country'].isin(['Mexico', 'Brazil', 'Canada']))
# Obtiene el nombre, username y edad de los usuarios que cumplen con la condición dada
df[condition][['name','username','age']]
```

```
# 23 Mostrar en consola el código postal de todos los usuarios de México.
df[df['country'] == 'Mexico']['postcode']
```

```
# 24 Obtener la edad que más se repite en el DataFrame.
df['age'].mode()
```

```
# 25 Obtener la edad que menos se repite en el DataFrame.
df['age'].value_counts().idxmin()
```
